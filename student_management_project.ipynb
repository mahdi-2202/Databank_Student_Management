# Project Explanation: Student Course Management System (Python)

## 1. Project Overview

This project implements a basic student course management system using Python. The program reads data from CSV files, stores it in memory, allows new records to be added, and retrieves information about which courses a student is enrolled in.


The system works with three datasets:
- students.csv – student information
- courses.csv – course information
- enrollments.csv – which students are enrolled in which courses


The program uses Python's csv module to read and write the data.

## 2. Import Modules and Data Structures

```python
import csv                 # imports the CSV module for working with CSV files.
import sys                 # imports the system module to interact with the Python runtime
```

Three lists store the data in memory:

```python
students = []      # [Student ID,First Name,Last Name,Age,Gender,Department,Year,GPA]
enrollments = []   # [Student ID,First Name,Last Name,Course ID]
courses = []       # [Course ID,Course Title,Department,Instructor,Credits,Semester]

# Each list represents a table loaded from a CSV file.
```

## 3. Main Functions

### addRecord()

```python
def addRecord(rec, data):
    if (len(rec) != len(data[0])):
        print("error")
        sys.exit()
    data.append(rec)
    
    
# Adds a new record to a table (students, courses, or enrollments).
```

### writeFile()

```python
def writeFile(fname, data):
    with open(fname, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile, delimiter=',', quotechar='|')
        for row in data:
            writer.writerow(row)
            
            
# Writes updated data back into a CSV file.
```

### readFile()

```python
def readFile(fname, data):
    with open(fname, newline='') as csvfile:
        for row in csv.reader(csvfile, delimiter=',', quotechar='|'):
            # ignore "deleted entries"
            if not(",".join(row).startswith("#")):  # Rows starting with # are ignored and They are treated as deleted records
                data.append(row)                    
                
                
# Reads records from a CSV file into memory.
```

### 4. Loading the Data

```python
# Loads all files into memory:

def readStudents():
    readFile('students.csv', students)

def readEnrollments():
    readFile('enrollments.csv', enrollments)

def readCourses():
    readFile('courses.csv', courses)

# read all data into memory
def initialLoading():
    readStudents()
    readEnrollments()
    readCourses()
```

### 5. Printing Records

```python
# This prints all records with their index.

def printRecords(table):
    for i,item in enumerate(table):
        print(i,item)

def printStudents():
    printRecords(students)

def printCourses():
    printRecords(courses)

def printEnrollments():
    printRecords(enrollments)
```

### 6. Finding Courses Taken by a Student

```python
def getCoursesTakenByStudentName(name):
    fname = name.split()[0]
    lname = name.split()[1]
    # identify student id
    studentId = -1
    for i,item in enumerate(students):
        if fname == item[1] and lname == item[2]:
            studentId = int(item[0])
    if (studentId < 0):
        print("error")
        return

    # (2) identify course ids
    course_ids = []
    for i,item in enumerate(enrollments[1:]):
        if int(item[0]) == studentId:
            course_ids.append(item[3])

    # (3) resolve course ids to course names
    taken_courses = []
    for i,item in enumerate(courses[1:]):
        for c in course_ids:
            if item[0] == c:
                taken_courses.append(item[1])
    return taken_courses


# Steps performed:

# Step 1 – Find the student ID: 
# Search the students table for matching first and last name.

# Step 2 – Find enrolled course IDs:
# Look in enrollments for records containing the student ID.

# Step 3 – Resolve course IDs:
# Match those IDs with courses table to get course names.

# example:
# getCoursesTakenByStudentName("John Smith")
# output: ['Mathematics', 'Physics', 'Programming']

```
